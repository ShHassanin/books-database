```python
conn= sqlite3.connect("Books.db")

url="https://books.toscrape.com/"
response= requests.get(url)
soup= BeautifulSoup(response.text,"html.parser")
links= soup.find('ul', class_="nav nav-list").find_all('a')
links.remove(links[0])


    
categoryID =0
for link in links:
    url2= url+link.get('href')
    category= link.get_text().strip()
    categoryID +=1
    response2= requests.get(url2)
    soup2= BeautifulSoup(response2.text,"html.parser")
    articles= soup2.find_all('article', attrs={'class':"product_pod"})
    print(categoryID , category)
    conn.execute('INSERT INTO Categories (categoryID , category) VALUES (?,?)',[categoryID , category])

    try:
        nextt= soup2.find('li', attrs={'class':"next"}).find('a')
        while nextt !=None:


            url3= url2.replace('index.html', nextt.get('href'))

            response3= requests.get(url3)
            soup3= BeautifulSoup(response3.text,"html.parser")
            articles2= soup3.find_all('article', attrs={'class':"product_pod"})
            for r in articles2:
                articles.append(r)
            print(url3 ,category)
            nextt= soup3.find('li', attrs={'class':"next"}).find('a')


    except AttributeError:
        print('none')

    for article in articles:
            price= float((article.find('p', attrs={'class':'price_color'}).get_text()).replace('Â£',''))
            bookName= article.find('h3').find('a').get('title')
            stars = article.find('p').get('class')[1]
            starsdic= {'One':1,'Two':2,'Three':3,"Four":4,'Five':5}
            rating= int(starsdic[stars])
            print(bookName,price,rating, category )

            conn.execute('INSERT INTO bOOKS (categoryID,BOOKnAME,PRICE,RATING) VALUES (?,?,?,?)',[categoryID ,bookName , price , rating])
            
conn.commit()
```
