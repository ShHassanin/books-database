### 1)sql query to get books that has rate >=3 and has 'Mr' in its name.


```python
conn.execute('SELECT *  FROM Books WHERE rating >=3 AND bookName LIKE "%Mr%"').fetchall()
```

### 2)sql query to get books that has rate >=3 or price > 20.


```python
conn.execute('SELECT * FROM Books  WHERE rating >=3  OR price >20').fetchall()
```

###  3)sql query to get books that has rate not 3.


```python
conn.execute('SELECT * FROM BOOKS WHERE rating != 3').fetchall()
```

### 4)sql query to get books that has price between 10 and 40 and has rate 3 or 4 or 1.

```python
conn.execute('SELECT * FROM BOOKS WHERE price BETWEEN 10 AND 40 AND (rating =3 or rating =4 or rating= 1)').fetchall()
```

```python
conn.execute('SELECT * FROM BOOKS WHERE price BETWEEN 10 AND 40 AND (rating in (3,4,1))').fetchall()
```

### 5)sql query to get the top 5 most expensive books .


```python
conn.execute('SELECT * FROM BOOKS ORDER BY price DESC LIMIT 5').fetchall()
```

###  6)sql query to get the 3rd 10 books in the books table order first by rate desc and then by price asc.


```python
conn.execute('SELECT * FROM BOOKS ORDER BY rating DESC, price LIMIT 10 offset 3').fetchall()
```

### 14)sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.


```python
conn.execute('select categories.category, books.bookName, books.price, books.rating from categories join books on categories.categoryID=books.categoryID').fetchall()
```

### 15)sql query to calculate how many books each rate has and have price between 20£ and 30£.


```python
conn.execute('select books.rating,bookname, count(books.bookName), books.price from categories join books on categories.categoryID=books.categoryID where price between 20 and 30 group by books.rating ').fetchall()
```

### 16)sql query to calculate how many books each category has having count > 10.

```python
conn.execute('select categories.category, count(books.bookName) from categories join books on categories.categoryID=books.categoryID group by categories.category having count(books.bookName)>10 ').fetchall()
```

### 17)sql query to get all books with category_name='Music' using subquery.

```python
conn.execute('select  * from books where categoryid=(select categoryid from categories where category="Music")').fetchall()
```
